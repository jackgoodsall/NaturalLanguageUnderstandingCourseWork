## Configure Paths

Set `MODEL_CHECKPOINT` to the saved `Solution B` checkpoint path. If the final model is stored on the cloud, download it first and point this variable at the local `.pt` file.


```python
from pathlib import Path

REPO_ROOT = Path.cwd()
MODEL_CHECKPOINT = REPO_ROOT / 'official_coursework' / 'local_artifacts' / 'B_REMOTE_FULL_001_best_model.pt'
INPUT_CSV = REPO_ROOT / 'official_coursework' / 'released_test_data' / 'test_data' / 'ED' / 'test.csv'
OUTPUT_CSV = REPO_ROOT / 'outputs' / 'submission' / 'b_predictions_debug.csv'
SUBMISSION_FILE = REPO_ROOT / 'outputs' / 'submission' / 'Group_52_B.csv'
THRESHOLD = 0.60

print('MODEL_CHECKPOINT =', MODEL_CHECKPOINT)
print('INPUT_CSV =', INPUT_CSV)
print('OUTPUT_CSV =', OUTPUT_CSV)
print('SUBMISSION_FILE =', SUBMISSION_FILE)
print('THRESHOLD =', THRESHOLD)
```

```python
import subprocess
import sys

cmd = [
    sys.executable,
    '-m', 'src.solution_b.evaluate',
    '--checkpoints', str(MODEL_CHECKPOINT),
    '--data', str(INPUT_CSV),
    '--device', 'auto',
    '--threshold', str(THRESHOLD),
    '--output', str(OUTPUT_CSV),
    '--submission', str(SUBMISSION_FILE),
]

print('Running command:')
print(' '.join(cmd))
result = subprocess.run(cmd, capture_output=True, text=True)
print(result.stdout)
if result.returncode != 0:
    print(result.stderr)
    raise RuntimeError(f'Solution B inference failed with exit code {result.returncode}')
```

```python
print('Detailed output exists:', OUTPUT_CSV.exists())
print('Submission file exists:', SUBMISSION_FILE.exists())
if SUBMISSION_FILE.exists():
    with open(SUBMISSION_FILE) as f:
        for _ in range(5):
            line = f.readline()
            if not line:
                break
            print(line.rstrip())
```
